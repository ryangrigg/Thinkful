```python
#The interest rate column has '%' symbols in it, and the loan length column has the string 'months' in it. 
#The FICO range is a categorical variable; to use it in our calculations, we need it to be numeric. 

#************************* We can use dataframe[column_name].values[0] to pull a value to test before implementing for the
#************************* entire column. 

#Let's take one of the values and map it out:
v1 = loansData['Loan.Length'].values[0]
v1
```

```python
v1 = v1.rstrip(' months') #this takes the cell and returns only the first 3 characters; all we need for the FICO Score
v1
```

```python
v1 = int(v1) #turn it into a int; good to go. 
v1
```

```python
#Change FICO SCORE: 1) use the lower number 2) int datatype 3) column called: FICO.Score
loansData['FICO.Score'] = loansData['FICO.Range'].map(lambda x: int(x[:3]))

#Change the interest rate: 1) remove % 2) convert to float so 3% would be 0.03 3)round to 3 decimal places
loansData['Interest.Rate'] = loansData['Interest.Rate'].map(lambda x: round(float(x.rstrip('%'))/100, 3))

#Change the loan length: 1) remove ' months' *note the space as well* 2) convert to int
loansData['Loan.Length'] = loansData['Loan.Length'].map(lambda x: int(x.rstrip(' months')))
```

## Linear Regression Analysis

In this last assignment we're going to define a linear model with two independent variables: FICO Score and Loan Amount. This will help us determine the dependent variable, Interest Rate. We know that our linear model is going to look something like this:

InterestRate = b + a1(FICOScore) + a2(LoanAmount)

We're going to use statsmodels to find the model coefficients b (which is the y-intercept), a1, and a2.

```python
import statsmodels.api as sm
```

```python
#extract columns we're interested in:
intrate = loansData['Interest.Rate']
loanamt = loansData['Amount.Requested']
fico = loansData['FICO.Score']
```

```python
type(intrate)
```

When we extract a column from a DataFrame, it's returned as a Series datatype. You want to reshape the data like this:

```python
# The dependent variable; we are trying to predict the interest rate based off of the other two columns
y = np.matrix(intrate).transpose()

# The independent variables shaped as columns
x1 = np.matrix(fico).transpose()
x2 = np.matrix(loanamt).transpose()

#Transposing the matrix is necessary for easier multiplication 
```

```python
#quick look at the data structure:
print y[0:2]
print ''
print x1[0:2]
print ''
print x2[0:2]
```

```python
#Now you want to put the two columns together to create an input matrix (with one column for each independent variable):
x = np.column_stack([x1,x2])
x
```

```python
#Now we create a linear model:
X = sm.add_constant(x)
model = sm.OLS(y,X)
f = model.fit()

#Output the results summary:
f.summary()
```
